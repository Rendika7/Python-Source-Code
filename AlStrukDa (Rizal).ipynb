```python
import numpy as np
from numpy import random
import pandas as pd
```

```python
z = random.randint(20, 100, size=(10000))
```

```python
y = random.randint(100, 190, size = (10000))
```

```python
import random
x = []
for i in range (10000):
    result_str = ''.join((random.choice('FM')))
    x.append(result_str)
```

```python
dataset = {'jenis kelamin': x, 'Tinggi': y, 'Berat': z}
df = pd.DataFrame(dataset)
df
```

```python
#seluruh warga ascending
warga = df['Tinggi']
arrwarga= np.array(warga)
arrwarga
def selectionSort(itemlist):
    n =len(itemlist)
    for i in range(n-1):
        minValueIndex = i
        for j in range (i+1, n):
            if itemlist[j]<itemlist[minValueIndex]:
                minValueIndex = j
        if minValueIndex != i:
            temp = itemlist[i]
            itemlist[i] = itemlist[minValueIndex]
            itemlist[minValueIndex]=temp
    
    return itemlist
arrsemua = selectionSort(arrwarga)
print(arrsemua)
```

```python
datawargaasc = {'jenis kelamin' : x, 'Tinggi' : arrsemua, 'Berat': z}
dwasc = pd.DataFrame(datawargaasc)
dwasc
```

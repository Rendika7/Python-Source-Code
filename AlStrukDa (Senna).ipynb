```python
import pandas as pd
import numpy as np
from numpy import random
```

# Membuat Data Frame 10ribu Data

```python
df = pd.DataFrame(np.random.randint(1,10000,size=(10000, 3)), columns=list(['berat badan' ,'tinggi badan', 'jenis kelamin'])) 
df
```

## Menentukan Isi Berat Badan

```python
df['berat badan']= np.random.randint(18, 80, size=(10000, 1)) 
df
```

## Menentukan Isi Tinggi Badan

```python
df['tinggi badan']= np.random.randint(110, 190, size=(10000, 1)) 
df
```

## Menentukan Jenis Kelamin

```python
kelamin = ['laki-laki', 'perempuan']
jeniskelamin = np.random.choice(kelamin, 10000)
jeniskelamin = jeniskelamin.tolist()

df["jenis kelamin"] = jeniskelamin
df
```

---

# 1.1 Mengurutkan Tinggi Badan dari Pendek - Tinggi

```python
ndf = np.array(df['tinggi badan'])

def selectionSort(itemlist):
    n = len(itemlist)
    for i in range(n-1):
        minValueIndex = i
        for j in range (i+1, n):
            if itemlist[j] < itemlist[minValueIndex]:
                minValueIndex = j
        if minValueIndex != i:
            temp = itemlist[i]
            itemlist[i] = itemlist[minValueIndex]
            itemlist[minValueIndex] = temp
    
    return itemlist
newVarPT = selectionSort(ndf)
# print(pd.DataFrame(newVarPT, columns = list(['tinggi badan'])))
```

```python
varPT = pd.DataFrame(newVarPT)
varPT.rename(columns={0:'tinggi badan'})
```

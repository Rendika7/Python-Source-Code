```python
MergeSortTime = [0.00474,0.01078,0.03088,0.06559]
Nama = ["(100 , 1000)","(200 , 2000)","(300 , 5000)","(500 , 10000)"]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(8,6))
plt.bar(Nama,MergeSortTime, color='blue')

font1 = {'family':'sans-serif','color':'Red','size':20, }
font2 = {'family':'sans-serif','color':'Blue','size':15}

plt.title('MERGE SORT', fontdict = font1)
plt.ylabel('TIMES', fontdict = font2)
plt.xlabel("(ITERATION , DATA)", fontdict = font2)
plt.xticks(size=8)
plt.yticks(size=8)
plt.grid(axis = 'y')
plt.show()
```

```python
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline

Nama = ["(100 , 1000)","(200 , 2000)","(300 , 5000)","(500 , 10000)"]

x = np.arange(len(Nama))
width = 0.25

fig, ax = plt.subplots(figsize=(20, 35))

mhsa = ax.bar(x + width*2, SelectionSortTime, width, label='Selection Sort', color = 'green')
mhsi = ax.bar(x + width, QuickSortTime, width, label='Quick Sort', color='purple')
mhsx = ax.bar(x, MergeSortTime, width, label='Merge Sort', color= "blue")


ax.grid(axis = "y")
ax.set_title('Time Complexity', size=20)
ax.set_ylabel('TIME', size=14)
ax.set_xlabel('(ITERATION, DATA)', size=16)
ax.set_xticks(x)
ax.set_xticklabels(Nama, size=12)
ax.legend(fontsize=14)
```

```python
def factorial(num):
  if num == 0 or num == 1 :
    return 1
  else : return num * factorial(num-1)

print(factorial(969))
```

```python
#2 Base Case
def countdown(i): # akan menjadi infinite loop
  print(i)
  countdown(i+1)

countdown(6)
```

```python
def countdown(i): # akan menjadi infinite loop
  print(i)
  if i <= 1 : # Base Casae
    return
  else : # Recursive Case
    countdown(i-1)

countdown(6)
```

```python
def fibonacci(x):
  if x<= 0 :
    return 0
  elif x==1 :
    return 1
  fib = fibonacci(x-1)+fibonacci(x-2)
  return fib

fibonacci(5)
```

```python
def fib(n):
  a = 0
  b = 1
  if n==1 :
    print(a)
  else :
    print(a)
    print(b)
    for i in range(2,n):
      c = a + b
      a = b
      b = c
      print(c)
fib(5)
```

```python
n = int(input("Enter number of terms: "))
n1, n2 = 0, 1 # first two terms of fibonacci series
i = 0  #variable for loop

if n <= 0:
  print("Please enter a positive integer")
elif n == 1:
  print("Fibonacci sequence upto",n,":")
  print(n1)
else:
  print("Fibonacci sequence upto:")
  for i in range(n):  #for loop upto nth term
      print(n1)
      sum = n1 + n2  #Logic for the fibonacci series
      n1 = n2
      n2 = sum
      i += 1
```

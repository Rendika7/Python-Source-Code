```python
Jumlah_Populasi = [10.5,2.8,2.3,2.4]
Nama_Kota = ["Jakarta","Surabaya","Bandung","Depok"]

import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(8,6))
plt.barh(Nama_Kota,Jumlah_Populasi, color='blue')

font1 = {'family':'sans-serif','color':'Red','size':20, }
font2 = {'family':'sans-serif','color':'Blue','size':15}

plt.title('POPULASI', fontdict = font1)
plt.ylabel('Jumlah Populasi (Juta)', fontdict = font2)
plt.xlabel("Nama Kota", fontdict = font2)
plt.xticks(size=8)
plt.yticks(size=8)
plt.grid(axis = 'y')
plt.show()
```

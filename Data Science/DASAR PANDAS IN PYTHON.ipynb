# Contoh syntax yang bisa dikerjakan dengan PANDAS

```python
#import library pandas pada workspace / kernel saat ini
import pandas as pd
```

#### Mengenal DataFrame di Pandas

```python
#DataFrame sederhana dengan Dictionary
data1 = pd.DataFrame({"Nama" : ["Budi","Anto","Mawar"],
             "Jenis Kelamin" : ["Laki","Laki","Perempuan"]})

data1
```

```python
#index pada DataFrame
data1 = pd.DataFrame({"Nama" : ["Budi","Anto","Mawar"],
             "Jenis Kelamin" : ["Laki","Laki","Perempuan"]},
                    index = ["B","A","M"])

data1
```

```python
#mengubah index dengan menggantikan kolom yang ada menjadi index itu sendiri
data1 = pd.DataFrame({"Nama" : ["Budi","Anto","Mawar"],
             "Jenis Kelamin" : ["Laki","Laki","Perempuan"]})

data1.set_index('Jenis Kelamin')
```

#### Mengenal Series di Pandas

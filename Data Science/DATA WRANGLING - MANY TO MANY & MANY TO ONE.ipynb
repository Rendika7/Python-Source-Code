```python
import pandas as pd

df1 = pd.DataFrame({"key" :["b","b","a","c","a","a","b"], "data1" : range(7)})
df2 = pd.DataFrame({"key" :["a","b","d"],"data2" : range(3)})
print(df1)
print(df2)
pd.merge(df1,df2)
```

```python
# many to one
df3 = pd.DataFrame({"key_x" :[2,2,4,1], "index3" : range(4)})
df4 = pd.DataFrame({"key_y" :[1,2,4],"index4" : range(3)})
print(df3)
print(df4)
pd.merge(df4,df3,how = "right")
```

```python
# Many To One

import pandas as pd
df3 = pd.DataFrame({"key" :["b","b","a","c","a","a","b"], "data1" : range(7)})
df4 = pd.DataFrame({"key" :["a","b","d"],"data2" : range(3)})
pd.merge(df3,df4,how = "right")
```

```python
# many to many
import pandas as pd
df3 = pd.DataFrame({"key" :["b","b","a","c","a","a","b"], "data1" : range(7)})
df4 = pd.DataFrame({"key" :["a","a","b","b","d"],"data2" : range(5)})
pd.merge(df3,df4,how = "right")
```

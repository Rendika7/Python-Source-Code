```python
replace_gender = {"P" : 1, "L" : 0}
df["Gender (X1)"] = df["Gender (X1)"].map(replace_gender)

replace_status = {"NORMAL" : 0, "GEMUK" : 1, "KURUS" : 2}
df["Status IMT (X2)"] = df["Status IMT (X2)"].map(replace_status)

replace_berkacamata = {"Ya" : 1, "Tidak" : 0}
df["Berkacamata (X3)"] = df["Berkacamata (X3)"].map(replace_berkacamata)

replace_pernahsakit = {"Ya" : 1, "Tidak" : 0}
df["Pernah Sakit (X4)"] = df["Pernah Sakit (X4)"].map(replace_pernahsakit)

replace_gangguanpsikis = {"Ya" : 1, "Tidak" : 0}
df["Gangguan Psikis (X5)"] = df["Gangguan Psikis (X5)"].map(replace_gangguanpsikis)

replace_aktifbertanya = {"Ya" : 1, "Tidak" : 0}
df["Aktif Bertanya (X6)"] = df["Aktif Bertanya (X6)"].map(replace_aktifbertanya)

replace_aktifmenjawab = {"Ya" : 1, "Tidak" : 0}
df["Aktif Menjawab (X7)"] = df["Aktif Menjawab (X7)"].map(replace_aktifmenjawab)

replace_mengerjakantugas = {"Semua" : 1, "sebagian" : 0}
df["Mengerjakan Tugas (X8)"] = df["Mengerjakan Tugas (X8)"].map(replace_mengerjakantugas)

replace_tertarikmateri = {"Ya" : 1, "Mungkin" : 0}
df["Tertarik Materi (X9)"] = df["Tertarik Materi (X9)"].map(replace_tertarikmateri)

replace_alokasijambelajar = {"LEBIH DARI 10 JAM" : 0, "KURANG DARI 5 JAM" : 1, "ANTARA 5 - 10 JAM" : 2}
df["Alokasi Jam Belajar (X10)"] = df["Alokasi Jam Belajar (X10)"].map(replace_alokasijambelajar)

replace_memilikireferensitambahan = {"Ada" : 1, "Tidak Ada" : 0}
df["Memiliki Referensi Tambahan(X11)"] = df["Memiliki Referensi Tambahan(X11)"].map(replace_memilikireferensitambahan)

replace_browsingdanyoutube = {"Ya" : 1, "Tidak" : 0}
df["Browsing dan Youtube (X12)"] = df["Browsing dan Youtube (X12)"].map(replace_browsingdanyoutube)

replace_mengulangmateri = {"Ya" : 1, "Kadang-kadang" : 0}
df["Mengulang Materi (X13)"] = df["Mengulang Materi (X13)"].map(replace_mengulangmateri)

replace_praktekmandiri = {"Ya" : 1, "Kadang-kadang" : 0}
df["Praktek Mandiri (X14)"] = df["Praktek Mandiri (X14)"].map(replace_praktekmandiri)

replace_berdiskusi = {"Ya" : 1, "Tidak" : 0, "Kadang-kadang" : 2}
df["Berdiskusi (X15)"] = df["Berdiskusi (X15)"].map(replace_berdiskusi)

replace_memilikiHP = {"Ya" : 1, "Tidak" : 0}
df["Memiliki HP(X16)"] = df["Memiliki HP(X16)"].map(replace_memilikiHP)

replace_memilikilaptop = {"Ya" : 1, "Tidak" : 0}
df["Memiliki Laptop (X17)"] = df["Memiliki Laptop (X17)"].map(replace_memilikilaptop)

replace_kecukupankuotainternet = {"Ya" : 1, "Tidak" : 0, "Kadang-kadang" : 2}
df["Kecukupan Kuota Internet (X18)"] = df["Kecukupan Kuota Internet (X18)"].map(replace_kecukupankuotainternet)

replace_dukungansuasanarumah = {"Ya" : 1, "Tidak" : 0, "Kadang-kadang" : 2}
df["Dukungan Suasana rumah (X19)"] = df["Dukungan Suasana rumah (X19)"].map(replace_dukungansuasanarumah)

replace_PLN = {"Ya" : 1, "Tidak" : 0}
df["PLN (X20)"] = df["PLN (X20)"].map(replace_PLN)

replace_lokasi = {"Perkotaan" : 1, "Pedesaan" : 0, "Pesisir" : 2}
df["Lokasi (X21)"] = df["Lokasi (X21)"].map(replace_lokasi)

replace_sinyal = {"Ya" : 1, "Tidak" : 0, "Sebagian" : 2}
df["Ketersediaan Sinyal  (X22)"] = df["Ketersediaan Sinyal  (X22)"].map(replace_sinyal)

df
```

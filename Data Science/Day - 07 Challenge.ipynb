- Buatlah kolom baru yang berisikan angka terurut dengan numpy.arange dan masukkan kedalam DataFrame kalian. Tampilkan!!
- Urutkan nilai data kolom berdasarkan kolom index 0 dan 1. simpan pada variabel baru. Tampilkan.
- Tampilkan rangking dari setiap kolom data
- Tampilkan rangkuman statistika deskriptif dari DataFrame

```python
from numpy import random
df3=  pd.DataFrame(random.randint(100,size=40).reshape(10,4), columns = ["Nilai","Umur","Berat","Tinggi"], index = ["Rendika","Sigit","Pikrie","Febrian","Thesion","Senna","Mita","Ryanta","Irma","Rizal"])
df3
```

```python
df3["Ukuran Sepatu"]= np.arange(10)
df3
```

```python
#Urutkan nilai data kolom berdasarkan kolom index 0 dan 1. simpan pada variabel baru. Tampilkan.

df4= df3.sort_values(by=["Nilai","Umur"])
df4
```

```python
df5= df4.rank
df5
```

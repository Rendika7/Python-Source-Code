1. Importkan file data_challenge1 dan data_challenge2 kedalam program dengan benar.

```python
import pandas as pd
filepath1= "FileDataSet_PakOby/data_challenge1.csv"
filepath2= "FileDataSet_PakOby/data_challenge2.csv"
df1= pd.read_csv(filepath1, sep = ";")
df2= pd.read_csv(filepath2, sep = ";")
```

3. buatlah dataframe baru yang berisikan 'Name', 'BetweennessCentrality', 'ClosenessCentrality', 'Degree' dan 'Stress'

```python
import numpy as np
df1_new_1= np.array([df1.name,df1.BetweennessCentrality,df1.ClosenessCentrality,df1.Degree,df1.Stress])
df1_new = pd.DataFrame(df1_new_1)
df1_new
```

```python
df1_new = pd.DataFrame(df1_new).T
df1_new
```

```python
df2_new=np.array([df2.name,df2.BetweennessCentrality,df2.ClosenessCentrality,df2.Degree,df2.Stress])
df2_new= pd.DataFrame(df2_new)
df2_new
```

```python
df2_new = pd. DataFrame(df2_new).T
df2_new
```

4. Gabungkan kedua dataset tersebut dengan benar (kolom sama)

```python
gabung_df = pd.concat([df1_new,df2_new],axis=1)
gabung_df
```

5. Simpan kedalam file csv/excel/JSON

```python
filepath_save= "FileDataSet_PakOby/Gabung_Challange_Pertemuan 9"
gabung_df.to_csv(filepath_save)
```

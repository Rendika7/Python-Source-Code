```python
with open("Scrapping Python/Latihan.html","r") as html_file :
    content = html_file.read()
    Soup = BeautifulSoup(content,"lxml")
    print(Soup.prettify())
```

```python
Favorite = Soup.find_all("div",class_= "My Favorite")
for tipe in Favorite :
    Favorite_Section = tipe.h1.text
#     print(Favorite_Section)
    print("-"*40)
    Content_Favorite = tipe.ul.text
#     print(Content_Favorite)
    print(f'My {Favorite_Section} is {Content_Favorite}')
```

```python
Favorite = Soup.find_all("div",class_= "My Favorite")
for tipe in Favorite :
    Favorite_Section = tipe.h1.text
    print("-"*40)
    Content_Favorite = tipe.ul.text.split()[np.random.randint(4)]
    print(f'My {Favorite_Section} is {Content_Favorite}')
```

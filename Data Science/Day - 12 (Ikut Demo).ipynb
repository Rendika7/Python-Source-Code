```python
link = "http://herbaldb.farmasi.ui.ac.id/v3/index.php?v=spesies"
r = requests.get(link)
```

```python
r
```

```python
soup = BeautifulSoup(r.text,"html.parser") #parser untuk mengambil data html nya
soup
```

```python
tabel = soup.find_all("table")[1]
tabel
```

```python
for i in tabel.find_all("tr"):
    print(i)
    spesies = i.find("td", style = "text-align:left;padding-left:10px;" )
```

```python
for i in tabel.find_all("tr"):
    for j in tabel.find_all("td"):
        print(j)
```

```python
nm = []
fm = []
pn = []
rf = []
for i in tabel.find_all("tr"):
    nama = i.find_all("td", style = "text-align:left;padding-left:10px;")[0].text
    family = i.find_all("td", style = "text-align:left;padding-left:10px;")[2].text
    penemu = i.find_all("td", style = "text-align:left;padding-left:10px;")[3].text
    referensi = i.find_all("td", style = "text-align:left;padding-left:10px;")[4].text
    nm.append(nama)
    fm.append(family)
    pn.append(penemu)
    rf.append(referensi)
```

```python
df = pd.DataFrame([nm,fm,pn,rf]).T
df
```

```python
df.to_csv("simpan_latihan.csv")
```

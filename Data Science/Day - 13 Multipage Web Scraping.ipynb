```python
import requests #harus import req library biar bisa mengambil data dari web online
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
```

```python
nm = []
fm = []
pn = []
ref = []
no = []
Id = []
var = []
stat = []
aks = []

for i in range(191) :
    Link = "http://herbaldb.farmasi.ui.ac.id/v3/index.php?v=spesies&sortby=&start=" + str(i)
    html_text = requests.get(Link)
    soup = BeautifulSoup(html_text.text,"html.parser")
    tabel = soup.find_all("table", border="0", width="100%")[1]
    for j in tabel.find_all("tr") :
        nomor = j.find_all("td", style = "text-align:center;")[0].text
        Id_Spesies = j.find_all("td", style = "text-align:center;")[1].text
        nama = j.find_all("td", style = "text-align:left;padding-left:10px;")[0].text
        varietas = j.find_all("td", style = "text-align:left;padding-left:10px;")[1].text
        family = j.find_all("td", style = "text-align:left;padding-left:10px;")[2].text
        penemu = j.find_all("td", style = "text-align:left;padding-left:10px;")[3].text
        refrensi = j.find_all("td", style = "text-align:left;padding-left:10px;")[4].text
        status = j.find_all("td", style = "text-align:left;padding-left:10px;")[5].text
        aksi = j.find_all("td", style = "text-align:left;padding-left:10px;")[6].text
        nm.append(nama)
        fm.append(family)
        pn.append(penemu)
        ref.append(refrensi)
        no.append(nomor)
        var.append(varietas)
        Id.append(Id_Spesies)
        stat.append(status)
        aks.append(aksi)
    print("Load Pada Page ke -",i)
```

```python
df = pd.DataFrame(np.array([no,Id,nm,var,fm,pn,ref,stat]), index = ["Nomor", "ID Spesies", "Spesies", "Varietas", "Family","Penemu", "Refrensi", "Status"])
df = df.T
df
```

```python
filepath_save= "FileDataSet_PakOby/Scrapping_Tanaman.csv"
df.to_csv(filepath_save,sep = ",")
```

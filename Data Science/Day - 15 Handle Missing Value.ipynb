```python
import numpy as np
import pandas as pd

#Menghilangkan batas jumlah maksimal tampilan kolom dari tabel pada Pandas
pd.set_option("display.max_row", None)
```

```python
#menggunakan parameter index_col = "Sample" agar dengan mudah mengubah columns menjadi kolom dari sample...
df = pd.read_csv("FileDataSet_PakOby/raw-material-properties.csv", index_col = "Sample")
df
```

```python
df.isnull().sum()
```

```python
df.isnull().sum(axis=1)
```

```python
# menghapus (nilai default nya ROW) yang mengandung missing value
df.dropna()
```

```python
# menggunakan parameter how yang dimana diberikan nilai "all" maka akan menghapus yang baris bernilai NaN semua dalam sebarisnya...
df.dropna(how="all")
```

```python
#menghapus baris yang memiliki missing value sebanyak 2 buah
df.dropna(thresh = 2)
```

```python
# Menghapus baris dari tabel dibawah, jika size3 dan density3 mengandung missing value
df.dropna(subset=['size3', 'density3'])
```

```python
df.fillna({"size1" : 0, 
           "size2" : "Satu", 
           "size3" : "Dua",
           "density1" : "Tiga",
           "density2" : "Empat",
           "density3" : "Lima"})
```

```python
#Mengisi missing value dengan mengcopy nilai yang ad sebelumnya. data teratas jika data kosong, maka akan tetap kosong karena tidak ada nilai sebelum data teratas
"""
1     1
NaN   1
NaN   1
4     4
"""
df.ffill()
```

```python
#Mengisi missing value dengan mengcopy nilai yang ada sesudahnya. data terbawah jika kosong maka akan tetap kosong
"""
1     1
NaN   4
NaN   4
4     4
"""
df.bfill()
```

```python
df.fillna({"size1" : df["size1"].mean(), 
           "size2" : df["size2"].mean(), 
           "size3" : df["size3"].mean(),
           "density1" : df["density1"].mean(),
           "density2" : df["density2"].mean(),
           "density3" : df["density3"].mean()})
```

```python
import pandas as pd
import numpy as np
import missingno as msno
```

```python
data = 'FileDataSet_PakOby/16. Dataset2-kamyr-digester.csv'
df = pd.read_csv(data, sep = ',')
pd.set_option("display.max_row", None)
pd.set_option("display.max_columns", None)
df
```

```python
msno.bar(df)
```

```python
Index = ['Observation', 'Y_Kappa', 'ChipRate','BF_CMratio', 'BlowFlow', 'ChipLevel4', 'T_upperExt_2',
            'T_lowerExt_2','UCZAA','WhiteFlow_4', 'AAWhiteSt_4', 'AA_Wood_4', 'ChipMoisture_4', 'SteamFlow_4', 
            'Lower_HeatT_3', 'Upper_HeatT_3', 'ChipMass_4', 'WeakLiquorF', 'BlackFlow_2', 'WeakWashF', 
            'SteamHeatF_3', 'T_Top_Chips_4', 'SulphidityL_4']

df.columns= Index
df
```

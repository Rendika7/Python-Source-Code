```python
import pandas as pd
import numpy as np
import missingno as msno
```

```python
#Menampilkan keseluruhan nilai kolom dan baris pada Dataframe dari csv yang diberikan.

data = 'FileDataSet_PakOby/16. Dataset1-travel-times.csv'
df = pd.read_csv(data, sep = ',')
pd.set_option("display.max_row", None)
pd.set_option("display.max_columns", None)
df
```

```python
# Mengecek berapa banyak baris dan kolom pada Dataframe

df.shape
```

```python
#Menampilkan Missing value dari Dataframe yang telah diberikan dengan berupa visualisasi bar plot, dikarenakan dari hasil penelitian diketahui pula bahwa manusia mampu mengingat 80% dari apa yang dilihat dan hanya mengingat 20% saja dari apa yang mereka baca. Dan menurut Wyzowl, sebanyak 8 dari 10 orang (atau 80%) didapati lebih mudah memahami sesuatu secara visual. jadi kelompok kami juga menyeddiakan visualisasi gambar

msno.bar(df)
#angka yang diatas adalah total banyaknya nilai yang tidak missing value
# yang sebelah kiri adalah presentasi terhadap banyaknya nilai yang tidak missing value, dimana 1.0 merepresentasikan 100%, dan 0.8 merepresentasikan 80% dan seterusnya
# yang dibawah menunjukkan nama nama kolom yang ada pada Dataframe
# Yang kanan menunjukkan jika nilai dari 1.0 adalah 205 berarti tidak ada missing value dalam kolom tersebut, jika 0.8 berarti 164, dan menunjukkan bahwa terdapat missing value pada kolom tersebut, dan seterusnya.
```

```python
# Melihat apakah pada setiap kolom ada missing value yang dimana berarti bernilai NaN atau None

df.isna().sum()
```

```python
# Setelah tahu tentang kolom mana yang mengandung missing value maka langsung kita cek, apakah kolom tersebut memiliki nilai yang tidak sesuai dengan nilai yang seharusnya, cara mengeceknya adalah menggunakan syntax pd.Unique(FuelEconomy), misal pada kejadian ini adalah semua nilai pada kolom FuelEconomy adalah bernilai angka, jika ada tidak bernilai angka maka kita harus handle nilai tersebut. ada dua cara yaitu pertama dengan menghapus atau me-replace dengan suatu nilai yang dimana biasanya adalah rata2 dari kolom tersebut dengan mengabaikan nilai yang NaN, lalu yang kedua dengan menghapusnya secara keseluruhan barisnya. penanganan ini juga berlaku jika kita ingin menghandle suatu nilai missing yang dimana biasanya bernilai NaN atau None

FuelEconomy = df["FuelEconomy"] 
pd.unique(FuelEconomy)
```

```python
# Setelah melakukan pengecekan tentang apakah ada nilai yang tidak seharusnya didalam suatu kolom FuelEconomy seperti codingan di atas, maka dilanjut dengan mencari keberadaan nilai yang tidak sesuai tersebut dengan menggunakan cara yang tidak manual tentunya karena nnti jika bekerja dengan banyak data pasti akan kesusahan jika menggunakan cara manual. disini yang akan dicari adalah nilai "-" yang ada pada kolom FuelEconomy. kenapa kita ingin menghapus nilai tersebut? karena nilai tersebut bukanlah nilai NaN atau None, "-" adalah nilai string, yang dimana akan menggangu kita ketika ingin mencari nilai rata2 dari kolom FuelEconomy nantinya jika diperlukan nilai rata-rata.
FuelEconomy = df[df["FuelEconomy"] == "-"]
FuelEconomy
```

```python
#Setelahnya ditemukan nilai yang tidak sesuai dengan kebutuhan, yaitu ada di baris 6 dan 7 pada Dataframe. maka langkah selanjutnya adalah dengan menghapus baris 6 dan 7 atau mereplacenya. disini keputusan yang kelompok kami ambil adalah dengan menghapus baris 6 dan 7 

#kenapa kelompok kami  menghapus baris tersebut? dikarenakan mising value pada bari 6 dan 7 tersebut adalah termasuk dalam jenis MNAR (Missing Not at Random) yaitu ada hubungan yang berbeda antara kecenderungan suatu nilai untuk hilang dan nilainya. value tersebut jika ada maka akan memiliki kesinambungan dengan kolom lainnya, jadi jika kami memberikan nilai yang dimana biasanya diambil dari rata2 kolom atau malah nilai bebas, itu dapat menganggu kesinambungan dalam 2 baris tersebut. jadi kami memutuskan untuk menghilangkan baris itu. karena FuelEconomy sendiri artinya adalah konsumsi bbm mobil setelah dipakai dalam jarak dan waktu tertentu.

#nah penting sekali untuk nilai2 pada FuelEconomy agar tetap berkesinambungan dengan nilai yang ada pada kolom-kolom lainnya, karena jika menganalisis dari FuelEconomy kita bisa mengetahui hal hal penting seperti contohnya, Jika FuelEconomy bernilai 9.08 itu berarti 9.08 liter digunakan per 100km, dikarenakan dataframe yang kita analisis kemungkinan besar tidak berasal dari indonesia dan kami berasumsi secara logis bahwa Dataframe ini berasal dari US (United Stated) yaitu dikarenakan di Indonesia tidak turun salju (dilihat dalam kolom Comments) maka nilai tersebut di convert kedalam nilai BBM di US maka menjadi Satuan MpG (Miles per Galon) dan menjadi  25.9047 U.S. mpg.

#dengan data yang telah diperoleh diatas maka dapat disimpulkan dalam satu baris yaitu baris ke index 16 yaitu penggunaan mobil pada tanggal 12/16/2011, hari jumat, jam 12:22, perjalanan ke rumah, dengan jarak 51.05 miles, dan top speed 132.2, dan total waktu berkendara adalah 35,2 memerlukan sebanyak 25.9047 U.S. mpg. untuk Fuel Economynya.

df_new = df.drop([6,7], axis=0, inplace = False)
df_new.shape
```

```python
#Pada bagian ini kami juga menghapus missing value pada FuelEconomy yang berupa nilai NaN.
df_new = df_new.dropna(subset=['FuelEconomy'])
index = np.arange(186)
df_new = df_new.set_index(pd.Index(index))
```

```python
#Setelah handling missing value untuk kolom FuelEconomy selesai maka dilanjut dengan menghandle kolom selanjutnya, jika dilihat pada keterangan di atas pada saat kita melakukan syntax df.isna().sum() maka selanjutnya adalah kolom "Comments". nah untuk kolom ini, kami mengutip dari beberapa sumber yaitu berkata demikian "If a column holds a lot of missing values, say more than 80%, and the feature is not meaningful, that time we can drop the entire column." jadi kamu memutuskan untuk menghapus kolom "Comments" ini.

del(df_new["Comments"])
```

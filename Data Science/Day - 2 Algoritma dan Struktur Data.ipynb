```python
def binary_search(arr, n):
    Awal = 0
    Akhir = len(arr) - 1
    Tengah = 0
    while Awal <= Akhir:
        Tengah = (Akhir + Awal) // 2
        if arr[Tengah] < n:
            Awal = Tengah + 1
        elif arr[Tengah] > n:
            Akhir = Tengah - 1
        else: 
            return Tengah
    return -1
arr = [3, 6 ,15, 19, 28, 36, 40, 41, 46, 56, 60, 71]
Data_Dicari = int(input("Masukkan angka yang ingin dicari : "))
r = binary_search(arr, Data_Dicari)
if r != -1:
    print("Element found at index", str(r))
else: 
    print("Element is not present in array")
```

```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
n=np.linspace(1,10,100)

big_o_cons= np.ones(n.shape)
big_o_log=np.log(n)
big_o_log_linear=n*big_o_log
big_o_quadratic=n**2
big_o_cubic=n**3
big_o_exponential=2**n
```

```python
labels=['Constant','Logarithmic','Linear','Log Linear','Quadratic','Qubic','Exponential']
big_o=[big_o_cons,big_o_log,big_o_log_linear,big_o_quadratic,big_o_cubic,big_o_exponential]
```

```python
plt.figure(figsize=(10,8))
plt.ylim(0,100)
for i in range(len(big_o)):
  plt.plot(n, big_o[i], label =labels[i])
plt.legend()
plt.ylabel('Runtime')
plt.xlabel('n')
```

```python
import pandas as pd
df = pd.DataFrame.from_dict({
    "Umur" : [10,35,34,23,70,55,89],
    "Tinggi" : [130,178,155,133,195,150,205],
    "Berat" : [80,100,100,120,90,87,111]
})
```

```python
df
```

## Normalize max-min

```python
def skala_min_max(series):
    return (series-series.min())/(series.max()-series.min())
```

```python
for col in df.columns:
    df[col] = skala_min_max(df[col])
print(df)
```

```python
import sklearn as sk
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
scaler.fit(df)
scaled = scaler.fit_transform(df)
scaled_df = pd.DataFrame(scaled, columns=df.columns)
scaled_df
```

## Normalisasi Standart (Z-score)

```python
def z_score(series):
    return (series-series.mean())/series.std()
for col in df.columns:
    df[col]= z_score(df[col])
df
```

## Absolute Max

```python
def absolute_max(series):
    return series/series.abs().max()
for col in df.columns:
    df[col]=absolute_max(df[col])
df
```

```python
Matrix = []
Baris = int(input("Masukkan Banyak Baris : "))
Kolom = int(input("Masukkan Banyak Kolom : "))

print("\nPencari Matrix")
for i in range (Baris) :
    New = []
    for y in range (Kolom) :
        New.append(int(input("Masukkan Angka |Baris,Kolom| ("+str(i+1)+","+str(y+1)+") : ")))
    Matrix.extend([New])

print("\nMatrix A \n")

for a in Matrix :
    for b in a :
        print(b, end="  ")
    print("")
print("---------------------")

print("Menu Pilihan : \n 1. \t Perkalian Skalar Pada Matrix \n 2. \t Determinan Matrix \n 3. \t Invers Matrix \n 4. \t Shape dari Matrix \n 5. \t Keluar")

#------------------------------------------------------------------------------------------------------ 

def Perkalian ():
    
#------------------------------------------------------------------------------------------------------ 
    Angka_Perkalian = int(input("Masukkan Angka Yang Ingin Dikalikan : "))
    print("Hasil Matrix A x",Angka_Perkalian)
    for baris in Matrix :
        for nilai in baris :
            print(nilai*Angka_Perkalian,end=" ")
        print("")
#------------------------------------------------------------------------------------------------------ 
        
def Determinan ():
    
#------------------------------------------------------------------------------------------------------ 
    if Baris == Kolom :
        if Baris == 1 and Kolom == 1 : #---Ordo 1X1---
            Det_1 = (Matrix[0][0])
            print(Det_1)
        elif Baris == 2 and Kolom == 2 : #---Ordo 2X2---
            Det_2 = (Matrix[0][0]*Matrix[1][1] - Matrix[0][1]*Matrix[1][0])
            print(Det_2)
        elif Baris == 3 and Kolom == 3 : #---Ordo 3X3---
            Det_3 = (Matrix[0][0]*Matrix[1][1]*Matrix[2][2]\
                    + Matrix[0][1]*Matrix[1][2]*Matrix[2][0]\
                    + Matrix[0][2]*Matrix[1][0]*Matrix[2][1]\
                    - Matrix[2][0]*Matrix[1][1]*Matrix[0][2]\
                    - Matrix[2][1]*Matrix[1][2]*Matrix[0][0]\
                    - Matrix[2][2]*Matrix[1][0]*Matrix[0][1])
            print(Det_3)
        elif Baris > 3 and Kolom > 3 :
            print("Maaf Tugasnya hanya sampai Matrix 3x3 saja")
    else : print("Tidak Ada Hasil Determinan (Determinan Harus Menggunakan Matrix Persegi)")
#------------------------------------------------------------------------------------------------------

def Inverse ():
    
#------------------------------------------------------------------------------------------------------    
    if Baris == Kolom :
        if Baris == 1 and Kolom == 1 : #---Ordo 1X1---
            print("Matrix 1x1 tidak mempunyai inverse Matrix")
        elif Baris == 2 and Kolom == 2 : #---Ordo 2X2---
            #Rumus DETERMINAN ORDO 2
            Det_2 = (Matrix[0][0]*Matrix[1][1] - Matrix[0][1]*Matrix[1][0])
            #Rumus INVERSE ORDO 2
            Inverse_2 = [[Matrix[1][1]/Det_2, -Matrix[0][1]/Det_2],[-Matrix[1][0]/Det_2, Matrix[0][0]/Det_2]]
            #Menapilkan Output dari Inverse ordo 2
            for i in Inverse_2 :
                for y in i :
                    print(y, end = " ")
                print("")
        elif Baris == 3 and Kolom == 3 : #---Ordo 3X3---
            #RUMUS DETERMINAN ORDO 3
            Det_3 = (Matrix[0][0]*Matrix[1][1]*Matrix[2][2]\
                    + Matrix[0][1]*Matrix[1][2]*Matrix[2][0]\
                    + Matrix[0][2]*Matrix[1][0]*Matrix[2][1]\
                    - Matrix[2][0]*Matrix[1][1]*Matrix[0][2]\
                    - Matrix[2][1]*Matrix[1][2]*Matrix[0][0]\
                    - Matrix[2][2]*Matrix[1][0]*Matrix[0][1])
            #RUMUS ADJOIN
            adj = [[Matrix[2][2]*Matrix[1][1]-Matrix[2][1]*Matrix[1][2],-(Matrix[2][2]*Matrix[0][1]-Matrix[2][1]*Matrix[0][2]),Matrix[1][2]*Matrix[0][1]-Matrix[1][1]*Matrix[0][2]],
       [-(Matrix[2][2]*Matrix[1][0]-Matrix[2][0]*Matrix[1][2]),Matrix[2][2]*Matrix[0][0]-Matrix[2][0]*Matrix[0][2],-(Matrix[1][2]*Matrix[0][0]-Matrix[1][0]*Matrix[0][2])],
       [Matrix[2][1]*Matrix[1][0]-Matrix[2][0]*Matrix[1][1],-(Matrix[2][1]*Matrix[0][0]-Matrix[2][0]*Matrix[0][1]),Matrix[1][1]*Matrix[0][0]-Matrix[1][0]*Matrix[0][1]]]
            print("A-1 = 1/Det A x adj A")
            print("    = 1/",Det_3, adj, "\n")
            #Rumus INVERSE ORDO 3 = setiap angka dalam adjoin dikalian dengan 1/Det_3. Yang dimana Det_3 adalah determinan dari variabel Matrix yang berordo 3
            for i in adj :
                for y in i :
                    print(y*(1/Det_3),end=" ")
                print("") 
        elif Baris > 3 and Kolom > 3 : #---Ordo Diatas 3---
            print("Maaf Tugasnya hanya sampai Matrix 3x3 saja")
    else : print("Tidak Ada Hasil Inverse (Inverse Harus Menggunakan Matrix Persegi, Karena diperlukan hasil Determinannya)")
#------------------------------------------------------------------------------------------------------ 
    
def Shape ():
    
#------------------------------------------------------------------------------------------------------ 
    if Baris == Kolom :
        if Baris == 1 and Kolom == 1 : #---Ordo 1X 1---
            print("Shape dari matrix A adalah : 1 x 1 = 1")
            #-------------Shape adalah hasil jumlah dari semua angka yang ada didalam matrix Tersebut
            Total_1x1 = Matrix[0][0]
            print("Total Seluruh Nilai =",Total_1x1)
        elif Baris == 2 and Kolom == 2 : #---Ordo 2X2---
            print("Shape dari matrix A adalah : 2 x 2 = 4")
            #-------------Shape adalah hasil jumlah dari semua angka yang ada didalam matrix Tersebut
            Total_2x2 = Matrix[0][0] + Matrix[0][1] + Matrix[1][0] + Matrix[1][1]
            print("Total Seluruh Nilai =",Total_2x2)
        elif Baris == 3 and Kolom == 3 : #---Ordo 3X3---
            print("Shape dari matrix adalah : 3 x 3 = 9")
            #-------------Shape adalah hasil jumlah dari semua angka yang ada didalam matrix Tersebut
            Total_3x3 = Matrix[0][0] + Matrix[0][1] + Matrix[0][2]\
                        + Matrix[1][0] + Matrix[1][1] + Matrix[1][2]\
                        + Matrix[2][0] + Matrix[2][1] + Matrix[2][2]
            print("Total Seluruh Nilai =",Total_3x3)
#------------------------------------------------------------------------------------------------------ 
    
    
#Input untuk user agar dapat memilih operasi apa yang akan dijalankan
Ulangi = "Ya"
while Ulangi == "Ya" :
    Pilihan_1 = int(input("Masukkan Pilihan Anda (1/2/3/4/5): "))
    if Pilihan_1 == 1 :
        Perkalian()
        print("")
    elif Pilihan_1 == 2 :
        Determinan()
        print("")
    elif Pilihan_1 == 3 :
        Inverse()
        print("")
    elif Pilihan_1 == 4 :
        Shape()
        print("")
    elif Pilihan_1 == 5 :
        Pilihan_2 = input("Anda yakin ingin keluar? (Ya/Tidak) --> ")
        if Pilihan_2 == "Ya" :
            Ulangi = "Tidak"
        else : 
            Ulangi = "Ya"
        print("")
    else :
        Print("Mohon Untuk Memasukkan Pilihan Dengan Benar !!")
```

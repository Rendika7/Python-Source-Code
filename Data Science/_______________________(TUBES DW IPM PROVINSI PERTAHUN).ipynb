```python
import numpy as np
# from numpy import mean
import pandas as pd
import matplotlib.pyplot as plt
# pd.set_option("display.max_row", None)
# pd.set_option("display.max_columns", None)
from statistics import mean
# pd.set_option("display.max_column", None)
df = pd.read_csv("TUBES Data WRANGLING/IPM-Provinsi-Indonesia.csv",sep=';')
df = df.set_index("Nama Provinsi")
```

```python
Duasatu = df.iloc[0:34,0:]
Duapuluh = df.iloc[34:68,0:]
Sembilanbelas = df.iloc[68:102,0:]
Delapanbelas = df.iloc[102:136,0:]
Tujuhbelas = df.iloc[136:170,0:]
Enambelas = df.iloc[170:204,0:]
```

```python
Nama_Provinsi = Duasatu.index.tolist()
```

```python
data = {'2016': IPM_2016,
        '2017' : IPM_2017,
        '2018': IPM_2018,
        '2019': IPM_2019,
        '2020': IPM_2020,
        '2021': IPM_2021,}


df = pd.DataFrame(data)
df['Provinsi'] = Nama_Provinsi
df = df.set_index('Provinsi')

df

df.plot(kind='bar', stacked=True,figsize = (21,7),color = ["indigo","orangered","magenta","springgreen","red","turquoise"],
        title='Index Pembangunan Manusia',xlabel = "Provinsi",ylabel = "Nilai")
```

```python
plt.figure(figsize=(21,7))
plt.xticks(rotation = 'vertical')
plt.grid(axis = 'y')

plt.title('IPM Tahun 2021', size=30, color = "red")
plt.xlabel('Nama Provinsi', size=20, color='purple')
plt.ylabel("Nilai IPM", size=20, color='purple')

plt.bar(Nama_Provinsi,IPM_2021, color='turquoise')
```

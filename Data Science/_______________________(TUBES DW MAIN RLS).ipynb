```python
import numpy as np
# from numpy import mean
import pandas as pd
import matplotlib.pyplot as plt

pd.set_option("display.max_row", None)
# pd.set_option("display.max_columns", None)
from statistics import mean
pd.set_option("display.max_column", None)
df = pd.read_excel("TUBES Data WRANGLING/IPM Tahun 2016 -2021.xlsx")
df
```

```python
df = df.drop(0, axis=0)
df = df.drop(549, axis=0)
df.columns = ["Provinsi/Kabupaten/Kota","IPM-2021","IPM-2020","IPM-2019","IPM-2018","IPM-2017","IPM-2016","UHH-2021","UHH-2020","UHH-2019","UHH-2018","UHH-2017","UHH-2016","HLS-2021","HLS-2020","HLS-2019","HLS-2018","HLS-2017","HLS-2016","RLS-2021","RLS-2020","RLS-2019","RLS-2018","RLS-2017","RLS-2016","PPK-2021","PPK-2020","PPK-2019","PPK-2018","PPK-2017","PPK-2016"]
df
```

```python
RLS_ = df[["RLS-2016","RLS-2017","RLS-2018","RLS-2019","RLS-2020","RLS-2021"]]
RLS_
```

# NORMALIZE DATA

```python
def skala_min_max(RLS):
    return (RLS-RLS.min())/(RLS.max()-RLS.min())

for col in RLS_.columns:
    RLS_[col] = skala_min_max(RLS_[col])
RLS_
```

```python
provinsi = df["Provinsi/Kabupaten/Kota"]
RLS_ = RLS_.set_index(provinsi)
RLS_
```

# SPLIT DATA per Daerah

# SUMATERA

```python
SUMATERA = RLS_.iloc[0:164,0:]
```

```python
SUMATERA = SUMATERA.drop(index='ACEH')
SUMATERA = SUMATERA.drop(index='SUMATERA UTARA')
SUMATERA = SUMATERA.drop(index='SUMATERA BARAT')
SUMATERA = SUMATERA.drop(index='RIAU')
SUMATERA = SUMATERA.drop(index='JAMBI')
SUMATERA = SUMATERA.drop(index='BENGKULU')
SUMATERA = SUMATERA.drop(index='LAMPUNG')
SUMATERA = SUMATERA.drop(index='KEP. BANGKA BELITUNG')
SUMATERA = SUMATERA.drop(index='KEPULAUAN RIAU')
```

```python
kolom_IPM = ["RLS-2016","RLS-2017","RLS-2018","RLS-2019","RLS-2020","RLS-2021"]
RSUMATERA = []
for i in kolom_IPM :
    Anj = SUMATERA[i]
    enambelas=[]
    for i in Anj :
    #     print(type(i))
        enambelas.append(i)
    f = round(mean(enambelas), 2)
    RSUMATERA.append(f)
RSUMATERA
```

```python
plt.plot(HLS_Grafik, label= ["Jawa","Sumatera","Sulawesi","Papua","Maluku","Kalimantan","Bali","Nusa Tenggara"])
plt.xlabel('TAHUN')
plt.ylabel('RATA-RATA')
plt.title("Rata-rata Lama Sekolah")
plt.rcParams["figure.figsize"] = (9,9)
plt.grid(True)
plt.legend();
plt.show
```

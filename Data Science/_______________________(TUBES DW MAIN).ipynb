```python
def list_split(listA, n):
    for x in range(0, len(listA), n):
        every_chunk = listA[x: n+x]

        if len(every_chunk) < n:
            every_chunk = every_chunk + \
                [None for y in range(n-len(every_chunk))]
        yield every_chunk

SetiapKota = list(list_split(kota, 6))
# print(SetiapKota)
print(SetiapKota[0])
```

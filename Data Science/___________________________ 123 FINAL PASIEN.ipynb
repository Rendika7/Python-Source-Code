```python
class pasien :
    def __init__(self) :
        self.pasien = []
        self.nomor_antrian = []
        global m
        m = []
    def tambah(self):
        antrian = len(self.nomor_antrian)
        antrian += 1
        nama = input('nama pasien: ')
        self.pasien.append(nama)
        m.append(antrian)
        self.nomor_antrian.append(antrian)
        print("nomor antrian anda : ",antrian)
        print("berhasil menambah antrian anda")
    def nomor_sekarang(self):
        if len(self.pasien) == 1 :
            print("pasien saat ini :", self.pasien[0])
            print("tidak ada pasien selanjutnya")
        elif len(self.pasien) > 1:
            print("pasien saat ini :", self.pasien[0])
            print("pasien selanjutnya :", self.pasien[1])
        else:
            print("BELUM ADA ANTRIAN")
    def hapus(self):
        self.pasien.pop(0)

data = pasien()
kesempatan = 3
p = 0
while p == 0:
    user = input('ID: ')
    pwd = input('Password: ')
    if (user == 'woy') & (pwd == 'apa'):
        print("Anda telah berhasil login")
        kondisi = 0
        while kondisi == 0:
            print('\n===============================================================================================')
            print('menu:')
            print('1. tambah antrian')
            print('2. lihat nomor sekarang')
            print('3. keluar')
            pick = int(input("Masukkan pilihan (1/2/3): "))
            print()
            if pick == 1:
                data.tambah()
            elif pick == 2:
                data.nomor_sekarang()
                if len(m) < 1 :
                    continue
                else:
                    selesai= input("apakah pasien saat ini sudah selesai?(y/n): ")
                    if selesai == 'y':
                        data.hapus()
                        continue 
            else:
                print(" ")
                print("SELESAI")
                break
        break   
    
    elif (user != 'woy') & (pwd != 'apa'):
        kesempatan -= 1
        if kesempatan > 0:    
            print("ID atau Password salah, silahkan coba lagi")
            print('kesempatan login sisa', kesempatan,'kali\n')
        elif kesempatan == 0:
            print("ID atau Password salah")
            print('========================COBA LAGI NANTI===============================')
            break
   
```

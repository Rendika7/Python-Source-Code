```python
import pandas as pd
import numpy as np
```

# Import 3 File CSV

```python
df1 = pd.read_csv("Data Set/android_devices_new.csv", sep = ";")
df1.isnull().sum()
```

```python
df2 = pd.read_csv("Data Set/user_device.csv", sep = ",")
df2.isnull().sum()
```

```python
df3 = pd.read_csv("Data Set/user_usage.csv", sep = ",")
df3.isnull().sum()
```

# Mengolah Kolom Device dan User_id

```python
pd.set_option("display.max_row", None) #Menampilkan seluruh row pada DataFrame
df1_2gabung = pd.merge(df1, df2,on = "device", how = 'outer') #Menggabung df1 dan df2 berdasarkan kolom device
Baru = df1_2gabung.dropna(subset = ["device","user_id"]) #Menghapus Device dan User_id yang Bernilai NaN
Indexing = np.arange(1,275) #Untuk membuat indexing baru agar lebih rapi
Baru = Baru.copy() # Mengcopy dataframe, karena jika tdiak dicopy akan melanggar aturan dari pengimport-an csv. yaitu menambahkan index berbeda dari originalnya.
Baru["Number"] = Indexing# Menambahkan kolom baru, yaitu kolom indexing yang berisi urutan angka dari 1 sampai 275
Baru = Baru.set_index("Number") #Menggunakan syntax set_index untuk mengubah sebuah kolom dari dataframe menjadi index dari dataframe itu sendiri
Baru # Menampilkan isi Dataframe
```

```python
#Mencari jenis jenis device yang unique
Device = Baru["device"].unique()
Device = Device.tolist()
for i in Device :
    print(i,",",end = "")
```

```python
Grafik = [] # list kosong untuk tempat dari jumlah masing2 merk
Nama = [] # list kosong untuk Nama merk
for i in Device : # perulangan digunakan untuk mengetahui banyaknya merk tanpa menggunakan metode manual
    Merk = Baru[Baru["device"] == i] # Mengakses setiap device yang bermerk sama yaitu i, yang dimana i adalah setiap value di kolom device unique diatas tadi.
    Jumlah_Merk = Merk["device"].tolist() # mengakses setiap merk yang bernilai sama, dan mengubahnya ke list.
    if len(Jumlah_Merk) >= 7 : # percabangan untuk memilah dan mengambil merk yang muncul lebih dari sama dengan 7 kali
        Grafik.append(len(Jumlah_Merk)) # menambahkan banyaknya setiap merk ke dalam list "Grafik"
        Nama.append(i) # menambahkan value i kedalam list Nama. yang dimana i bernilai nama device yang unique tadi
        print("Jumlah device bermerk", i ,":",len(Jumlah_Merk)) # Menampilkan seluruh info dari merk yang berbeda dengan syarat memiliki jumlah lebih dari sama dengan 7
print("Ini list dari Grafik :",Grafik) # menampilakan jumlah munculnya suatu merk yang lebih dari sama dengan 7
print("Ini list dari Nama :",Nama) # menampilkan nama - nama device yang jumlahnya lebih dari sama dengan 7
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(8,6))
plt.bar(Nama,Grafik, color='Black')

font1 = {'family':'sans-serif','color':'Red','size':20, }
font2 = {'family':'sans-serif','color':'Orange','size':15}
font3 = {'family':'sans-serif','color':'Green','size':15}

plt.title('DATA DEVICE', fontdict = font1)
plt.ylabel('Jumlah', fontdict = font2)
plt.xlabel("Device", fontdict = font3)
plt.xticks(size=8)
plt.yticks(size=8)
plt.grid(axis = 'y')
plt.show()
```

### Menemukan banyak Pengguna Ios dan Android

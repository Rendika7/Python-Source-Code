```python
def partition(arr, low, high):
  i = (low-1)
  pivot = arr[high]

  for j in range(low, high):
    if arr[j] <= pivot:
      i += 1
      arr[i], arr[j] = arr[j], arr[i]
  arr[i+1], arr[high] = arr[high], arr[i+1]
  return (i+1)

def quickSort(arr,low,high):
  if len(arr) ==1:
    return arr
  if low < high:
    pi = partition(arr,low,high)
    quickSort(arr,low,pi-1)
    quickSort(arr,pi+1,high)
```

```python
import sys
sys.setrecursionlimit(10000000)
import time
import numpy as np
from statistics import mean
```

```python
Array_Pertama_QuickSort = np.random.randint(101, size = 1000)

Time = []
for i in range(100):
    start = time.time()
    n = len(Array_Pertama_QuickSort)
    quickSort(Array_Pertama_QuickSort,0,n-1)
    end = time.time()
    waktu = end-start
    Time.append(waktu)
print(Time)
AVG = mean(Time)
AVG_100 = round(AVG, 5)
print("100 times Iteration with 1000 data, the average running time of the program is:", AVG_100)
```
